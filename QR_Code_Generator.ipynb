### Running the QR code generator

##### The myqr.run() function is used to generate the QR code. It takes two important arguments:

The first argument is the URL or text that you want to encode into the QR code. In this case, it encodes my GitHub link.
The second argument is the file name for the output image.

The code will generate a QR code with the content https://github.com/kalravanshika and save it as GithubQR.png.

```python
myqr.run('https://github.com/kalravanshika', save_name='GithubQR.png')
```

### Displaying the QR Code in Jupyter Notebook

```python
from PIL import Image
```

```python
# Load and display the QR code image
img = Image.open('GithubQR.png')
display(img)
```

## Customization of the QR Code Design:

Adding a background image behind the QR code for a personalized look

```python
myqr.run(words = 'https://github.com/kalravanshika', picture='picture.png', colorized=True, save_name='qr_with_background.png')
```

### Displaying the Customized QR Code in Jupyter Notebook

```python
img_bg = Image.open('qr_with_background.png')
display(img_bg)
```
